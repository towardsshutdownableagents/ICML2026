```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
path = "/Users/carissacullen/IPP/plots/data/metrics_overtime/PPO/DREST/test_usefulness"
metric_list = []
for file in os.listdir(path):
    df = pd.read_csv(f'{path}/{file}')
    df = df.fillna(0)
    custom_step = df["custom_step"]
    metric = df[df.columns[-1]]
    metric_list.append(metric)
    plt.plot(custom_step,metric,c="blue",alpha=0.3)
metric_mean = np.mean(np.stack(metric_list),axis=0)
plt.plot(custom_step,metric_mean,c="blue")
```

```python
metric_sd = np.std(np.stack(metric_list),axis=0) 

print(f'${metric_mean[-1]:.3f} \pm {metric_sd[-1]:.3f}$')
```
